```python
import pandas as pd
```

```python
dataloader = pd.read_csv("/content/CSE299pr.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
dataloader.info()
```

```python
dataloader.isnull().sum()
```

```python
df = dataloader
new_df = df.dropna()
```

```python
new_df.isnull().sum()
```

```python
label_encoder = LabelEncoder()
```

```python
new_df.head()
```

```python
new_df['Weather_Type'] = label_encoder.fit_transform(new_df['Weather_Type'])
```

```python
new_df.head(20)
```

```python
# Columns to normalize
columns_to_normalize = ['Minimum_Temperature', 'Maximum_Temperature', 'Temperature', 'Dew_Point', 'Relative_Humidity', 'Wind_Speed', 'Wind_Direction', 'Precipitation', 'Visibility', 'Cloud_Cover', 'Sea_Level_Pressure']

# Min-Max scaling
scaler = MinMaxScaler()
new_df[columns_to_normalize] = scaler.fit_transform(new_df[columns_to_normalize])
print(new_df[columns_to_normalize])
```

```python
X = new_df.drop(['Weather_Type'], axis=1)
X
```

```python
Y = new_df['Weather_Type']
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
svm_classifier = SVC()
```

```python
svm_classifier.fit(X_train, y_train)
```

```python
y_pred = svm_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier()
```

```python
# Train the Random Forest classifier
rf_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Make predictions on the test set
predictions = rf_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
#Run decision tree model
from sklearn.tree import DecisionTreeClassifier
```

```python
# Creating a decision tree classifier
clf = DecisionTreeClassifier()
```

```python
# Training the classifier on the training set
clf.fit(X_train, y_train)
```

```python
# Making predictions on the testing set
y_pred_clf = clf.predict(X_test)
```

```python
# Evaluating the classifier's performance
accuracy = accuracy_score(y_test, y_pred_clf)
print("Accuracy:", accuracy)
```

```python
#KNN model
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
# Training the classifier on the training set
knn.fit(X_train, y_train)

# Making predictions on the testing set
y_pred_knn = knn.predict(X_test)

# Evaluating the classifier's performance
accuracy = accuracy_score(y_test, y_pred_knn)
print("Accuracy:", accuracy)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
# Creating a logistic regression model
lr = LogisticRegression()

# Training the model on the training set
lr.fit(X_train, y_train)

# Making predictions on the testing set
y_pred_lr = lr.predict(X_test)

# Evaluating the model's performance
accuracy = accuracy_score(y_test, y_pred_lr)
print("Accuracy:", accuracy)
```

```python
#cross validation with logistic regression
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression


# Initialize the classifier
classifier = LogisticRegression()

# Define the cross-validation strategy
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
cv_scores = cross_val_score(classifier, X, Y, cv=kfold, scoring='accuracy')

# Print the cross-validation scores
print("Cross-validation scores:", cv_scores)

# Print the mean and standard deviation of the cross-validation scores
print("Mean CV accuracy:", cv_scores.mean())
print("Standard deviation of CV accuracy:", cv_scores.std())
```

```python
#cross validation with random forest(highest accuracy)

# Define the cross-validation strategy
kfold = KFold(n_splits=4, shuffle=True, random_state=42)

# Perform cross-validation
cv_scores = cross_val_score(rf_classifier, X, Y, cv=kfold, scoring='accuracy')

# Print the cross-validation scores
print("Cross-validation scores:", cv_scores)

# Print the mean and standard deviation of the cross-validation scores
print("Mean CV accuracy:", cv_scores.mean())
print("Standard deviation of CV accuracy:", cv_scores.std())
```
